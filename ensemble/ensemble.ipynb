```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics

def myMetrics(y_pred, y_test, thres = 0.5):
    
    Fpr, tpr, thresholds = metrics.roc_curve(y_test, y_pred)
    roc_auc = metrics.auc(Fpr,tpr)
    print("AUC",  roc_auc)
    
    
    
    y_pred = [1 if x > thres else 0 for x in y_pred]
    
    
    cm= confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()
    precision=tp/(tp+fp)
    recall=tp/(tp+fn)
    fpr = fp/(fp+ tn)
    accuracy = (tp + tn)/(tn + tp + fn + fp)
    F1 = 2 * (precision * recall) / (precision + recall)
    print("precision:", precision*100)
    print("recall:", recall*100)
    print("false positive rate:", fpr*100)
    print("accuracy",  accuracy*100)
    print("F1-score", F1)
    
    
    return Fpr, tpr
```

```python
fpr, tpr = myMetrics(tests.xgb, tests.label, thres = 0.7)
plt.plot(fpr,tpr, linewidth = 1, c = 'r', label = 'xgb')


plt.legend(loc=4)
fpr, tpr = myMetrics(tests.log, tests.label, thres = 0.8)
plt.plot(fpr,tpr, linewidth = 1, c = 'y', label = 'log')
plt.legend(loc=4)
fpr, tpr = myMetrics(tests.ann, tests.label, thres = 0.996)
plt.plot(fpr,tpr, linewidth = 1, c = 'b', label = 'ann')
plt.legend(loc=4)
fpr, tpr = myMetrics(tests.rf, tests.label, thres = 0.8)
plt.plot(fpr,tpr, linewidth = 1, c = 'g', label = 'rf')
plt.legend(loc=4)
plt.title('ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```
